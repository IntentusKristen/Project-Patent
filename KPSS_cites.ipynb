```python
import pandas as pd
```

```python
cite_df = pd.read_csv("finished_df_5.csv")
KPSS_df = pd.read_csv("KPSS_cpc_clean.csv")
```

```python
# merge data frame with KPSS data to get permno for each patent
KPSS_df.rename(columns={"patent_num": "patent_id"}, inplace=True)
merged_df = KPSS_df.merge(cite_df, on="patent_id", how="inner")
merged_df["num_of_patent_by_year"] = merged_df.groupby(["permno", "gyear"])["patent_id"].transform("nunique")

merged_df
```

```python
# isolate the number of patents a company has by year and company to get the cumulative total number of patents
cumulative_num_of_patents = merged_df[["permno", "gyear", "num_of_patent_by_year"]]
cumulative_num_of_patents.drop_duplicates(inplace=True)

# get the culmulative occurrence of number of patents
cumulative_num_of_patents.sort_values(by="gyear", inplace=True)
cumulative_num_of_patents["cumulative_patent_count"] = cumulative_num_of_patents.groupby("permno")["num_of_patent_by_year"].cumsum()
cumulative_num_of_patents[cumulative_num_of_patents["permno"] == 12060]

merged_df = merged_df.merge(cumulative_num_of_patents, on=["permno", "gyear"], how="inner")
merged_df
```

```python

merged_df["tech_prop_iota"] = (merged_df["revised_tech_prop"]**(1/1.5))

merged_df
```

```python
# calculate the patent distance by using given formula
# permno-patent-year level, for each permo patent, sum up the tech prop 
sum_tech_prop = merged_df
sum_tech_prop["tech_prop_sum"] = sum_tech_prop.groupby(["permno", "patent_id", "gyear"])["tech_prop_iota"].transform("sum")

sum_tech_prop["patent_dist_066"] = ((1/merged_df["cumulative_patent_count"])*(merged_df["tech_prop_sum"]))**(1.5)
sum_tech_prop
```
